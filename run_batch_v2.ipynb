# V2 batch simulation runner -- rich JSONL logs

Each deck cell runs the full matrix for that **P1 deck**:
5 P1 agents x 5 P2 agents x 9 P2 decks x 20 games = **4,500 games** (~1-2 h),
writing one `.jsonl` per game into `log_v2/<agent1>_<deck1>/` plus a `summary.csv`.

* **Resumable**: finished games are skipped, so you can interrupt the kernel and re-run a cell.
* Run cells in any order; each deck is independent. All 9 cells = 40,500 games (~10-15 h, ~6 GB).
* Log format: see the docstring of `sim_rich.py` (game_meta / mulligan / decision / game_end records; per-decision legal options, chosen action, full both-sides state snapshot, attributed events).
* **Decks are the restored 30-card lists from `decks_v2.py`** (all implemented cards put back; Zoo padded with a 2nd Abusive Sergeant + Crazed Alchemist for the unimplemented Dark Peddler). Not directly comparable to the old `log/` runs.
* `search_secs` in the records is for debugging only -- do **not** use it as a behavioral metric.

## Progress overview
Re-run anytime to see how far the batch has come.

```python
import csv, glob, os
root = v2.DEFAULT_LOG_ROOT
total_done = total_err = 0
print("%-40s %8s %8s %8s" % ("directory", "games", "errors", "of 900"))
for sp in sorted(glob.glob(os.path.join(root, "*", "summary.csv"))):
    with open(sp, newline="", encoding="utf-8") as f:
        rows = list(csv.DictReader(f))
    err = sum(1 for r in rows if r["winner"] == "ERROR")
    ok = len(rows) - err
    total_done += ok; total_err += err
    print("%-40s %8d %8d %8s" % (os.path.basename(os.path.dirname(sp)), ok, err,
          "done" if ok >= 900 else ""))
print("\nTOTAL: %d games logged, %d errors  (target 40,500)" % (total_done, total_err))
```
